```python
data4 = pd.read_csv("C:\\Users\\EMZ\\Desktop\\JeelAIDM\\NLP\\Data4.csv",header=None)
data4.columns=["email"]
data4
```

```python
for row in data4["email"]:
    valid_email = re.findall(r'^[a-z 0-9]+[\._?][a-z 0-9]+[@]\w+[.]+\w+$',row)
    if valid_email:
        print(valid_email)
```
